```python
# Import the packages in this project
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

#  Read in the data
df = pd.read_csv(r"C:\Users\DELL\Desktop\movies.csv")
```

```python
# Take a look at the data
df.head()
```

```python
# See if we have any missing data
# Loop through the data and see if there is anything missing
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df = df.dropna()
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for columns

print(df.dtypes)
```

```python
# Change Data Types of columns
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
print(df.dtypes)
```

```python
# Create correct Year column (The year in 'released' is different from the year in 'year')
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df.head()
```

```python
df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
pd.set_option('display.max_rows',100)
```

```python
df.drop_duplicates()
```

```python
#Budget high correlation
#Company high correlation
```

```python
#Scatter plot with budget vs gross
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
#Plot budget vs gross using seaborn
sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
# Look at correlation
df.corr(method='pearson') #pearson kendall spearman
```

```python
#Correlation Matrix
correlation_matrix=df.corr(method='pearson')
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Flim Features')
plt.ylabel('Flim Features')
sns.heatmap(correlation_matrix,annot=True)
plt.show()
```

```python
df_numerized=df
for col_name in df_numerized:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
#Correlation Matrix for Numerized Data
correlation_matrix=df_numerized.corr(method='pearson')
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Flim Features')
plt.ylabel('Flim Features')
sns.heatmap(correlation_matrix,annot=True)
plt.show()
```
